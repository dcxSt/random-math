```python
import numpy as np
import math
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0,4), ylim=(-2,2))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([],[])
    return line,

def animate(i):
    x = np.linspace(0,4,1000)
    y = np.sin(2*np.pi*(x-0.01 * i))
    line.set_data(x,y)
    return line,

anim = FuncAnimation(fig,animate,init_func=init,
                     frames=200,interval=20,blit=True)

anim.save('sine_wave.gif',writer='imagemagick')
```

```python
def exp_term(x,n):
    return(x**n / math.factorial(n))

# function that calculates first k terms for a given x, returns sequence
def get_k_terms_weight(x,k=100):
    return [exp_term(x,n) for n in range(k)] / np.exp(x)

def gauss(x,mu,sigma):
    return 1 / np.sqrt(np.pi*sigma**2)*e**(((x-mu) / sigma)**2)
```

```python
k=1000
x=5
for x in range(400,951,5):
    # plot the k'th first terms for a given x and their relative weights
    fig = plt.figure(figsize=(8,8))
    x_plt = [i for i in range(k)]
    y = get_k_terms_weight(x,k)
    mean = np.dot(x_plt,y)
    
    area,count = 0,0
    while area < .33:
        area+= y[int(mean)+count]
        count+=1
    std=count# just a rough estimate
    
    plt.bar(x_plt,y,label="x = {}\nmean = {}\nstd = {}".format(x,mean,std))
    plt.legend()
    plt.savefig('exp_{}.png'.format(x))
```
