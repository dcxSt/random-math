```python
# Collatz Reloaded
import array as arr
import matplotlib.pyplot as plt
import numpy as np
```

```python
# takes input starting number, calculates and displayes the collatz sequence
# returns the number of iterations that happened, and other variables - count
def calculate_collatz(n=100,display=True):
    if display: print("Collatz n=",n)
    count,count_even,count_odd = 0,0,0
    while n!= 1:
        count+=1
        if n%2==0:
            count_even+=1
            n/=2
        else:
            n*=3
            n+=1
            count_odd+=1
        n=int(n)
        if display:
            print(n,end="  ")
    return count,count_even,count_odd
```

```python
calculate_collatz(n=100)
```

```python
k=100000

x = [i for i in range(1,k)]
y = []

for i in x:
    y.append(calculate_collatz(i,display=False))
yt = np.transpose(y)
```

```python
# find the ratio of evens and odds, showld converge sto something as the numbers get bigger...
# first we do them in order of n then we do them in order of collatz length
ratios = []
for i in y:
    if i[1]!= 0: ratios.append(i[2]/i[1])# odd divided by even
    else: ratios.append(0)
```

```python
plt.figure(figsize=(19,19))

# plot the collatz number verses the number
plt.subplot(2,2,1)
plt.plot(x,yt[0],"bx")
plt.title("og collatz sequence")
plt.ylabel("collatz number")
plt.xlabel("n")
plt.grid()

# plot the ratio of even odd, this should converge to something like 
plt.subplot(2,2,2)
plt.plot(x,ratios,"rx")
plt.ylabel("ratio")
plt.xlabel("n")
plt.title("odd divided by even")
plt.grid()

# plot the collatz number and the ratio
plt.subplot(2,2,3)
plt.plot(yt[0],[yt[2][i]/yt[1][i] for i in range(len(yt[1]))], "gx")
plt.title("ratio by collatz number")
plt.ylabel("ratio, odd divided by even")
plt.xlabel("collatz number")
plt.grid()

### save the figure
#plt.savefig("collatz_plot_1.png")
```
